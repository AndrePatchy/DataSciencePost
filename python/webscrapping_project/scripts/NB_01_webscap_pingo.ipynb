TimeoutException wasn't used in this code, but since this is like the 7th version I didn't notice and imported anyway. finally, Beautifull Soup to extract the information one wants from the HTML structure. 

```python
import time 
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from bs4 import BeautifulSoup
```

### Mercearia

```python
# the driver will automatically open directely in the page of the category mercearia.

driver = webdriver.Chrome("path/to/chromedriver")
driver.get("https://mercadao.pt/store/pingo-doce/category/mercearia-11?page=0")

# It will wait for the COOKIE and accept it 
try:
    cookie_consent_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "onetrust-accept-btn-handler"))
    )
    cookie_consent_button.click()
except:
    pass

# sleep 
time.sleep(10)

# timestamp
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# empty list called results to store... wait for it... the results
results = []

# Yes, I counted how many pages there were available on "VER TODOS" of the category mercearia and looped them 
for i in range(0, 37):

    # print statement to check the state of the scraping process
    print(f"Going to page {i}")
    
    # NEXT PAGE (the first iteration will be page 0)
    driver.get(f'https://mercadao.pt/store/pingo-doce/category/mercearia-11?page={i}')
    
    # sleep again...xiu....
    time.sleep(20)
    
    # get the html source to the html object and go sleep!
    html = driver.page_source
    time.sleep(20)
    
    # Here one prepares the soup, sleep, and use attribute the find all method to the object products to use. Sleep again. 
    soup = BeautifulSoup(html, 'lxml')
    time.sleep(5)
    products = soup.find_all('pdo-product-item', class_='P9eg53AkHYfXRP7gt5njS ng-star-inserted')
    time.sleep(5)

    # Soup prepared. Time to scrape what one wants. In this case, everything that is available! Name, Quantity and prices. 
    for product in products:

        product_name = product.find('h3', class_ = 'pdo-heading-s detail-title').text

        product_quantity = product.find('pdo-product-price-per-unit', class_ = 'pdo-block').text
            
        price_per_unit = product.find('h4', class_ = 'detail-price').text

        # Store the results in the result dictionary, thinking ahed on the pandas data frame that is going to be build later. 
        result = {
            'Name': product_name,
            'Quantity': product_quantity,
            'Price': price_per_unit,
            'Time scraped': timestamp
                }
        results.append(result)
    
    # print statements to check the state of the scraping process
    print(f"Products scraped: {result}")
    print(f'Page {i} scraped')
    print(f"Current results list: {results}")    
    
    #WAIT 
    time.sleep(30)

driver.quit()
```

```python
mercearia = results
print(len(mercearia))
```

## Concatenate results and exporting to a csv file

```python
data_pd = mercearia + results_frutas_legumes + results_talho + results_peixaria + results_charcutaria + results_padaria + results_take_away + results_congelados + results_leitinhos + results_fridge
print(len(data_pd))
```

```python
import csv

file = 'pingo_doce_data.csv'

header = ['Name', 'Quantity', 'Price', 'Time scraped']

data = [[p['Name'], p['Quantity'], p['Price'], p['Time scraped']] for p in data_pd]

with open(file, 'w', newline = '') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)
```
