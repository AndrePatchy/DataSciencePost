### Concatenate the data sets in the intersected columns and respecting the indexes.

```python
continente = pd.read_csv(r'conti.csv',  encoding='utf-8')
pingo_doce = pd.read_csv(r'pingo_doce.csv',  encoding='utf-8')
```

```python
## add column Store to dataset
continente['Store'] = 'continete'
pingo_doce['Store'] = 'pingo_doce'
```

```python
## rename column
continente = continente.rename(columns={"Unnamed: 0": "Old_index",})
```

```python
## increment value of Index
pingo_doce['Old_index'] = pingo_doce['Unnamed: 0'] + len(continente)
```

```python
intersec_columns = []

for column in continente.columns:
  if column in pingo_doce.columns:
    intersec_columns.append(column)

intersec_columns
```

```python
continente_concat = continente[intersec_columns]

pingo_doce_concat = pingo_doce[intersec_columns]

frames = [continente_concat, pingo_doce_concat]

df_concat = pd.concat(frames)

df_concat = df_concat.reset_index()
df_concat = df_concat.drop(['index'], axis=1)

df_concat
```

Standerdization of categories

```python

category_mapping = {
    'mercearia': ['mercearia', 'bio-e-escolhas-alimentares'],
    'padaria_pastelaria': ['padaria_pastelaria', 'padaria-e-pastelaria'],
    'frutas_legumes': ['frutas_legumes', 'frutas-e-legumes'],
    'peixaria_talho': ['talho', 'peixaria', 'peixaria-e-talho'],
    'leite_ovos': ['leite_ovos_natas', 'laticinios-e-ovos'],
    'congelados': ['congelados'],
    'frigorifico': ['take-away', 'frigorífico', 'refeicoes-faceis', 'charcutaria-e-queijos']
}

conditions = [
    df_concat['Category'].isin(category_mapping['mercearia']),
    df_concat['Category'].isin(category_mapping['padaria_pastelaria']),
    df_concat['Category'].isin(category_mapping['frutas_legumes']),
    df_concat['Category'].isin(category_mapping['peixaria_talho']),
    df_concat['Category'].isin(category_mapping['leite_ovos']),
    df_concat['Category'].isin(category_mapping['congelados']),
    df_concat['Category'].isin(category_mapping['frigorifico'])
]

choices = [
    'mercearia',
    'padaria_pastelaria',
    'frutas_legumes',
    'peixaria_talho',
    'leite_ovos',
    'congelados',
    'frigorifico'
]

df_concat['Standardized_Category'] = np.select(conditions, choices, default=df_concat['Category'])
```

```python
category_mapping = {
    'MERC': ['mercearia', 'bio-e-escolhas-alimentares'],
    'PAPA': ['padaria_pastelaria', 'padaria-e-pastelaria'],
    'FRLE': ['frutas_legumes', 'frutas-e-legumes'],
    'PETA': ['talho', 'peixaria', 'peixaria-e-talho'],
    'LEOV': ['leite_ovos_natas', 'laticinios-e-ovos'],
    'CONGE': ['congelados'],
    'FRIGO': ['take-away', 'frigorífico', 'refeicoes-faceis', 'charcutaria-e-queijos']
}

conditions = [
    df_concat['Category'].isin(category_mapping['MERC']),
    df_concat['Category'].isin(category_mapping['PAPA']),
    df_concat['Category'].isin(category_mapping['FRLE']),
    df_concat['Category'].isin(category_mapping['PETA']),
    df_concat['Category'].isin(category_mapping['LEOV']),
    df_concat['Category'].isin(category_mapping['CONGE']),
    df_concat['Category'].isin(category_mapping['FRIGO'])
]

choices = [
    'MERC',
    'PAPA',
    'FRLE',
    'PETA',
    'LEOV',
    'CONGE',
    'FRIGO'
]

df_concat['Abbr_Category'] = np.select(conditions, choices, default=df_concat['Category'])
```

```python
df_concat.drop(['Category'], axis = 1, inplace = True)
```

### First filter. We filtered the first time with 1 token of each foodstuff (alimento) as one can see through the "basic_cabaz" list!

```python
## selecting just the products with name of basic cabaz
```

```python
#basic_cabaz = ['Leite', 'Queijo', 'Pão', 'Arroz', 'Massa', 'Cereal', 'Tosta', 'Bolacha', 'Feijão', 'Grão-de-bico', 'Ervilhas', 
   #            'Frango', 'Pescada', 'Atum', 'Sardinha', 'Cavala', 'Tomate', 'Vegetais', 'Brocolos', 'Espinafre', 'Cenoura', 'Alho', 'Azeite', 'Creme Vegetal', 'Marmelada']

basic_cabaz = [
    'Leite', 'Queijo', 'Arroz', 'Massa', 'Cereal', 'Tosta', 'Bolacha Maria', 
    'Feijão', 'Grão de bico', 'Ervilha', 'Frango', 'Pescada', 'Atum', 'Sardinha',
    'Cavala', 'Tomate', 'Vegetais', 'Brócolos', 'Feijão Verde', 'Espinafre', 'Cenoura',
    'Alho Francês', 'Azeite', 'Creme Vegetal', 'Marmelada'
]
```

```python
# Create an empty DataFrame
all_filtered_data = pd.DataFrame()

for index, basic_product in enumerate(basic_cabaz):
    pattern = r'{}'.format(basic_product)
    filtered = df_concat[df_concat['Name'].str.contains(pattern, flags=re.IGNORECASE)].copy()
    filtered.loc[:, 'Cabaz_product'] = index

    # Remove the values already present in all_filtered_data
    if len(all_filtered_data) > 0:
        filtered_df = filtered[~filtered['Old_index'].isin(all_filtered_data['Old_index'])]
    else:
        filtered_df = filtered

    all_filtered_data = pd.concat([all_filtered_data, filtered_df])

df_all_filtered_data = all_filtered_data.sort_index() 
```

## Second filter. Here we filter the "df_all_filtered_data" dataset that contains 3629 rows × 9 columns using rapidfuzz. Now we use all tokens of the foodstuff (alimentos) to improve the accuracy of the filter.  


Reducing the df_all_filtered_data with the rapidfuzz library

```python
items_cabaz = [
    'Leite Meio Gordo', 'Queijo Meio Gordo', 'Arroz', 'Massa', 'Cereais', 'Tostas', 'Bolacha Maria',
    'Feijão lata', 'Grão-de-bico lata', 'Ervilhas lata', 'Frango inteiro congelado',
    'Pescada congelada', 'Atum lata', 'Sardinha', 'Cavala lata', 'Tomate pelado',
    'Mistura de vegetais para sopa', 'Brócolos', 'Feijão Verde', 'Espinafres', 'Cenoura',
    'Alho Francês', 'Azeite', 'Creme Vegetal', 'Marmelada'
]
```

```python
df_re = df_all_filtered_data.copy()
```

```python
!pip install rapidfuzz
```

```python
# The code matches each product name of the full dataset with each of the name of the items_cabaz list and "build" a new data frame with the matched products. The WRatio was the scorer that present better results and the limit 15 is a reasonable number after testing.

from rapidfuzz import fuzz, process, utils

matches_data = pd.DataFrame(columns = df_re.columns)

for i in items_cabaz: 
  matches = matches = process.extract(i, df_re['Name'], scorer = fuzz.WRatio, limit = 15, processor = utils.default_process)
  for matched in matches:
    row = df_re.loc[df_re['Name'] == matched[0]]
    matches_data = pd.concat([matches_data, row])
  
matches_data.reset_index(drop = True, inplace = True)
```
